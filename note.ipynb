## Imports & Loading Datasets

```python
import pandas as pd
```

```python
# read excel file and convert to dataframe
df = pd.read_excel('supermarkt_sales.xlsx',
    # engine='openpyxl',
    sheet_name='Sales',
    skiprows=3,
    usecols='B:R',
    nrows=1000)
```

## Sales by Product ðŸ“Š

```python
df.groupby(by=["Product line"]).sum()
```

```python
df.groupby(by=["Product line"]).sum()[["Total"]]
```

```python
df.groupby(by=["Product line"]).sum()[["Total"]].sort_values(by="Total")
```

## Convert/ Add Hour Column

```python
df.info()
```

```python
# create an hour column
df['hour'] = pd.to_datetime(df['Time'], format="%H:%M:%S").dt.hour
```
